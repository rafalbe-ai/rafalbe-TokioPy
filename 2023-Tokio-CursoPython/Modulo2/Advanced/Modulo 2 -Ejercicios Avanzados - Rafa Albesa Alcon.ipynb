````python
#Importamos el módulo random.
import random
def pide_algo_por_pantalla(texto : str, tipo : str, entrada_min: float, entrada_max: float, lista_enumerada : list) -> str:
    '''Función que controla la introducción de parámetros por el teclado
        Entradas: 
            texto: str --> Texto a representar
            tipo: str --> Texto que indica el tipo de entrada ('entero', 'real' o 'enumerado' (lista de valores))
            entrada_min: float --> Valor mínimo para los tipos  'entero' y 'real'
            entrada_maxima : float --> Valor máximo para los tipos 'entero' y 'real'
            enumerado : list --> lista de valores válidos para el tipo 'enumerado'''
            
    while True:
        try:
            cadena = input(texto)
            if tipo == 'entero' and not(entrada_min <= float(cadena) <= entrada_max):
                print(f"El valor númerico debe estar entre {entrada_min} y {entrada_max}")
                cadena = str(int(cadena))
                mierror = True
            elif tipo == 'real' and not(entrada_min <= float(cadena) <= entrada_max):
                print(f"El valor númerico debe estar entre {entrada_min} y {entrada_max}")
                cadena = str(round(float(cadena),2))
                mierror = True
            elif tipo == 'enumerado' and cadena not in lista_enumerada:    
                print(f"El valor númerico debe estar uno de los siguientes {lista_enumerada}")
                mierror = True
            else:
                break
            
            if mierror: 
                mierror = False
                print(f"Error en la introduccón de datos. mierror = {mierror}")
            else:
                mierror = False
                print(f"El valor de salida de la cadena es: {cadena}")
        except ValueError:
            print("Tiene que introducir un valor válido.")
    return  cadena

def mostrar_alimentos_disponibles(diccionario_comidas):
    for comida, alimentos in diccionario_comidas.items():
        print(f"Los alimentmos disponibles para el {comida} son:")
        for alimento in alimentos['alimentos']:
            print(f"      - {alimento['nombre']} que tiene {alimento['calorias']} calorías y pertenece a las categorías de {alimento['categoria']}")    

def añadir_alimento(diccionario_alimentos : dict):
    #Preguntams si se desea introducir algún alimento nuevo.
    cadena = pide_algo_por_pantalla("¿Desea añadir algún alimento adicional a la lista de los alimentos? (s/n)", 
                                    tipo = 'enumerado', entrada_min = 0, entrada_max = 0, lista_enumerada=['s', 'n', 'S', 'N'])
    #si la respuesta es afirmativa, procedemos. Una vez se contesta afirmativamente, ya no hay marcha atrás. Se debe añadir un alimento.
    if cadena in ['s', 'S']:
        #Pedimos la comida para la que se desea añadir el alimento.
        cadena = pide_algo_por_pantalla("Introduzca la comida para la que desa añadir el alimento (1.- Desayuno, 2.- comida ó 3.- Cena)", 
                                        tipo = 'entero', entrada_min = 1, entrada_max = 3, lista_enumerada=['s', 'n', 'S', 'N'])
        #En función de la respuesta, ponemos el nombre de la comida en una variable temporal.
        if cadena == '1':
            comidaparaañadiralimento = 'desayuno'  
        elif cadena == '2':
            comidaparaañadiralimento = 'comida'  
        else:
            comidaparaañadiralimento = 'cena'  
        print(f"Va a añadir un alimento para la {comidaparaañadiralimento}.")
        #Pedimos el nombre del alimento que se desea añadir.
        nombre = input("Introduzca el nombre del alimento: ")
        #Pedimos en número de calorías que tiene el alimento que estamos añadiendo.
        calorias = pide_algo_por_pantalla("Introduzca el número de calorías del alimento",
                                            tipo = 'entero', entrada_min = 0, entrada_max = 10000, lista_enumerada=[])
        #Pedimos que se introduzca la categoría a la que pretenede el alimento.
        categoria = pide_algo_por_pantalla("Introduzca la categoría del alimento (Proteinas, Lácteos, Cereales, Vegetales): ", 
                                            tipo = 'enumerado', entrada_min = 0, entrada_max = 10000, lista_enumerada = ['Proteinas','Lácteos', 'Cereales', 'Vegetales'])
        #Con los datos obtenidos, poblamos el diccionario añadiendo el nuevo alimento..
        diccionario_alimentos[comidaparaañadiralimento]['alimentos'].append({'nombre' : nombre, 'calorias' : calorias, 'categoria' : categoria})
        
def generar_menu(diccionario_comidas : dict):
    '''Función que recibe un diccionario con el listado de comidas válidas por tipo de comida y devuelve un diccionario con el menú.
        Entradas: Diccionario que contiene la lista de comidas (puede ser una copia o el original, ya que sólo lo consulta, no lo modifica)
        Salida: La función devuelve una lista con tres elementos:
            Un diccionario que contiene el menú completo con las restriciones siguientes:
                1.- Tres platos por comida.
                2.- La suma de las calorías no puede superar las calorías máximas por comida. Si las supera, la función generará un menú de sólo dos alimentos
                para esa comida. Si aún así también la supera, genera un menú con un solo aliimento para esa comida. Obligamos a comer en todas las comidas.
            Un diccionario con las calorías que contiene cada una de las comidas
            Un diccionario con las calorías máximas permitidas por comida.
    '''
    
    #Generamos una lista con los tres valores límite de calorías para cada una de las comidas (desayuno, comida y cena)
    numcals_max_desayuno = float(pide_algo_por_pantalla("¿Cuántas calorías diarias quiere consumir como máximo en el desayuno", 
                            tipo = 'real', entrada_min = 200.0, entrada_max = 3000.0, lista_enumerada = []))
    numcals_max_comida = float(pide_algo_por_pantalla("¿Cuántas calorías diarias quiere consumir como máximo en cadala comida", 
                            tipo = 'real', entrada_min = 200.0, entrada_max = 3000.0, lista_enumerada = []))
    numcals_max_cena = float(pide_algo_por_pantalla("¿Cuántas calorías diarias quiere consumir como máximo en la cena", 
                            tipo = 'real', entrada_min = 200.0, entrada_max = 3000.0, lista_enumerada = []))
    #Generamos dos diccionarios, uno con las calorías máximas por comida y otro con para ir acumulando las calorías. 
    calorias_max = {'desayuno' : numcals_max_desayuno, 'comida' : numcals_max_comida, 'cena': numcals_max_cena}
    calorias = {'desayuno' : 0, 'comida': 0, 'cena': 0}
    #Inicializamos el diccionario que contendrá el menú. Es el diccionario que devolverá la función.'''
    menu_completo = {}
    #Iteramos por los distintos tipos de comida (desayuno, comida y cena) del diccionario de alimentos permitidos para cada comida.
    listacomidas = list(diccionario_comidas.keys())
    for indice, comida in enumerate(list(diccionario_comidas.keys())):
        #Generamos una lista con los índices de los elementos que componen los alimentos permitidos para la comida
        lista=list(range(len(diccionario_comidas[comida]['alimentos'])))
        #Generamos una lista  con tres números aleatorios distintos (utilizamos la función sample del módulo random.)
        tresaleatorios = random.sample(lista,3) #sorted(random.sample(lista, 3))
        #Calculamos las calorías que suman los alimentos seleccionados aleatoriamente.
        menu_completo[comida] = [diccionario_comidas[comida]['alimentos'][tresaleatorios[0]], diccionario_comidas[comida]['alimentos'][tresaleatorios[1]], diccionario_comidas[comida]['alimentos'][tresaleatorios[2]]]
        calorias[comida] = int(diccionario_comidas[comida]['alimentos'][tresaleatorios[0]]['calorias']) + int(diccionario_comidas[comida]['alimentos'][tresaleatorios[1]]['calorias']) + int(diccionario_comidas[comida]['alimentos'][tresaleatorios[2]]['calorias'])
        #Si nos hemos pasado de las catergorías límite, nos qudamos solo con dos alimentos de los tres que se había seleccionado aleatoriamente. 
        if calorias[comida] > calorias_max[comida]:
            menu_completo[comida] = [diccionario_comidas[comida]['alimentos'][tresaleatorios[0]], diccionario_comidas[comida]['alimentos'][tresaleatorios[1]]]
            calorias[comida] = diccionario_comidas[comida]['alimentos'][tresaleatorios[0]]['calorias'] + diccionario_comidas[comida]['alimentos'][tresaleatorios[1]]['calorias']
        #Si aún con dos alimentos nos pasamos del límite de calorías, nos quedamos únicamente con un alimento. Por lo menos, se tiene que ingerir un alimento por comida.
        if calorias[comida] > calorias_max[comida]:
            menu_completo[comida] = [diccionario_comidas[comida]['alimentos'][tresaleatorios[0]]]
            calorias[comida] = diccionario_comidas[comida]['alimentos'][tresaleatorios[0]]['calorias']
    #Devolvemos el menú generado junto con los diccionarios que contienen las calorías de cada comida y el máximo de calorías por comida
    return [menu_completo, calorias, calorias_max]


#Comidas válidas iniciales según el enunciado del problema.
desayuno = [
    {"nombre": "Leche desnatada", "calorias": 50, "categoria": "Lácteos"},
    {"nombre": "Yogur natural", "calorias": 70, "categoria": "Lácteos"},
    {"nombre": "Huevos revueltos", "calorias": 200, "categoria": "Proteínas"},
    {"nombre": "Tostadas de pan integral", "calorias": 100, "categoria": "Cereales"},
    {"nombre": "Fruta fresca", "calorias": 50, "categoria": "Vegetales"}
]

comida = [
    {"nombre": "Pollo asado", "calorias": 300, "categoria": "Proteínas"},
    {"nombre": "Pescado a la plancha", "calorias": 250, "categoria": "Proteínas"},
    {"nombre": "Ensalada de atún", "calorias": 200, "categoria": "Vegetales"},
    {"nombre": "Arroz integral", "calorias": 150, "categoria": "Cereales"},
    {"nombre": "Judías verdes", "calorias": 50, "categoria": "Vegetales"}
]

cena = [
    {"nombre": "Sopa de verduras", "calorias": 100, "categoria": "Vegetales"},
    {"nombre": "Ensalada mixta", "calorias": 150, "categoria": "Vegetales"},
    {"nombre": "Tortilla de patatas", "calorias": 250, "categoria": "Proteínas"},
    {"nombre": "Queso fresco", "calorias": 100, "categoria": "Lácteos"},
    {"nombre": "Frutos secos", "calorias": 150, "categoria": "Vegetales"}
]


#A partir de las listas de comidas creamos un diccionario de comidas.
''' Diccionario con la lista de comidas válidas:
    Cada comida es una lista de alimentos, siendo cada alimento un diccionario que tiene las siguientes clave valor:
        - 'nombre' : String que contiene el nombre del alimento.
        - 'calorias': entero que contiene las calorías del alimento.
        - 'categoría' : String que contiene la categoría del alimento de entre ('Proteinas', 'Lácteos', 'Cereales', 'Vegetales') '''

diccionario_comidas = {
    'desayuno': {'alimentos': desayuno},
    'comida': {'alimentos': comida},
    'cena': {'alimentos': cena},
}
    
    
    
    
    
    
'''Menú que controla la plicación.'''
while True: 
    #Pedimos por patnalla que se seleccione una de las tres opciones disponibles.
    seleccion = pide_algo_por_pantalla("""Introduzca la opción deseada:\n 
                            1.- Ver alimentos disponibles \n 
                            2.- Añadir un alimento\n 
                            3.- Mostrar un menú personalizado
                            99.- Para salir.... """, tipo = 'enumerado', entrada_min = 1.0, entrada_max = 3.0, lista_enumerada = ['1', '2', '3', '99'])
    #Si se selecciona la opción 1.- Ver alimentos disponibles.
    if seleccion == '1':
        #Mostramos los alimentos dispponibles por pantalla
        mostrar_alimentos_disponibles(diccionario_comidas)      
    #Si se selcciona la opcion 2.- Añadir un alimento
    elif seleccion == '2':
        #Llamada a la función que controla la introducción de un nuevo alimento en el diccionario.
        añadir_alimento(diccionario_comidas)
    #Si se elige la opción 3.- Generar un menú completo
    elif seleccion == '3':
        print('Generamos un menú personalizado según las características establecidas')
        #Llamada a la función que genera el menú.
        menu_completo = generar_menu(diccionario_comidas)
        #Imprimimos el resultado en pantalla
        for comida, alimentos in menu_completo[0].items():
            print(f"{comida.title()} con un total de {menu_completo[1][comida]} calorias (máximo de {menu_completo[2][comida]}):") 
            for alimento in alimentos:
                print(f"\t- {alimento['nombre']} ({alimento['calorias']} calorias)")        
    #Para salir del menú que controla la aplicación hay que introducir un 99.
    elif seleccion == '99':
        print("Salimos del programa. Hasta la próxima.")
        break

````
