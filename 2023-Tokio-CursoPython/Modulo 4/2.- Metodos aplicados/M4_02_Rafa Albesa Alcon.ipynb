**2) Solicita al usuario un texto y una palabra a buscar dentro del texto. Si la localiza, extraer desde esa posición hasta el final de la frase y almacenarlo en otra variable.** 
* Ejemplo de texto: Hola mundo que tal
* Buscamos: mundo 
* Resultado: mundo que tal

```python
cadena = input("Introduzca una cadena, por favor:")
palabra = input("Introduzca la palabra a buscar en dicha cadena:")
#print(f"La palabra '{palabra}'{'' if cadena.find(palabra) != -1 else ' no'} está contenida en la cadena '{cadena}'.")
indice = cadena.find(palabra)
if indice != -1:
    print(f"La palabra '{palabra}' está contenida en la cadena '{cadena}'")
    cadenaSalida = cadena[indice:]
    print(f"El resto de la cadena, a partir de la coincidencia, es {cadenaSalida}")
else:
    print(f"La palabra '{palabra}' no está contenida en la cadena '{cadena}'")


```
