```python
# Completa el ejercicio aquí
import numpy as np
matriz1 = np.arange(10,101)
print(matriz1)
matriz2 = np.flip(matriz1, 0)
print(matriz2)
matriz3 = np.arange(0,9).reshape(3,3)
print(matriz3)
identidad6x6 = np.identity(6)
print(identidad6x6)
aleatoria3x3x3 = np.random.randint(1, 1000, size = [3,3,3])
print(aleatoria3x3x3)
A = np.random.randint(0, 1000, size = (10,10))
B = np.random.randint(0, 1000, size = (10,10))
#Opcion de comparación 1 - Función integrada en numpy
print(f"Las matrices A y B {"" if np.array_equal(A, B) == True else "no"} son iguales")
#Opción de comparación 2: utilizando comparación elemento a elemento.
print(np.all(A == B, where = True))
```

```python
# Completa el ejercicio aquí
import pandas as pd
df_titulos = pd.read_csv("05_05_imdb_titulos.csv")
df_elenco = pd.read_csv("05_05_imdb_elenco.csv")
```

```python
print(df_titulos.head())
print(df_elenco.head())
print(f"El dataframe titulos tiene {df_titulos.shape[0]} filas y {df_titulos.shape[1]} columnas")
print(f"El dataframe elenco tiene {df_elenco.shape[0]} filas y {df_elenco.shape[1]} columnas")
#Usamos print (aunque queda peor) para que se quede en el terminal.
print(df_titulos.sort_values(by=['year']).head(5))
#Listamos los elementos de df_titulos que contienen la palabra "Dracula" en el título de la película
print(df_titulos[df_titulos['title'].str.contains('Dracula')==True])
#Mostramos cuántas películas contienen Drácula en su título (cuidado con el acento)
print(f"Existen {df_titulos[df_titulos['title'].str.contains('Dracula')==True].shape[0]} películas sobre Drácula en el dataframe")

#Genereamos una lista con los títulos ordenados por el número de veces que se repiten
listaTitulosOrdenadaPorRepeticiones = df_titulos['title'].value_counts().index.to_list()
#Mostramos los 10 primeros títulos de la lista
print(listaTitulosOrdenadaPorRepeticiones[:10])

#Ordenamos por fecha las películas que tienen como título 'Romeo and Juliet':
df1 = df_titulos[df_titulos['title'] == ('Romeo and Juliet')].sort_values(by=['year'])
#Nos quedamos con el valor del dataframe filtrado que tiene la menor fecha
print(f"La película buscada es {df1['title'][df1['year'] == df1['year'].min()].to_list()[0]} y es del año {df1['year'][df1['year'] == df1['year'].min()].to_list()[0]}")

#Ordenamos por fecha las películas que tienen como título 'Exorcist' ordenadas por fecha de menor a mayor:
print(df_titulos[df_titulos['title'].str.contains('Exorcist') == True].sort_values(by=['year']))

#Películas hechas en el año 1950
print(df_titulos[df_titulos['year'] == 1950])

#películas hechas entre 1950 y 1959 ambos inclusive.
print(df_titulos[((df_titulos['year'] >= 1950) & (df_titulos['year']<= 1959))].sort_values(by=['year']))


```

```python
#Mostrar todos los roles o papeles que hubo en la pelicula "The Godfather". También mostrar el número total de coincidencias 
#Se muestran por orden alfabético
print(df_elenco[df_elenco['title'] == 'The Godfather'].sort_values(by = ['name']))
print(f"El número de actores y roles es de {df_elenco[df_elenco['title'] == 'The Godfather'].sort_values(by = ['name']).shape[0]} actores")

# Mostrar el elenco completo ordenado por la clasificacion "n" de la pelicula "Dracula" de 1958
print(df_elenco[((df_elenco['title'] == 'Dracula') & (df_elenco['year'] == 1958))].sort_values(by = ['n']))

#Mostrar cuantos papeles de "Bruce Wayne" han sido hechos en la historia de las peliculas
print(df_elenco[((df_elenco['character'] == 'Bruce Wayne'))].sort_values(by = ['title']))

#Mostrar cuantos papeles ha hecho "Robert De Niro" en su carrera
print(df_elenco[((df_elenco['name'] == 'Robert De Niro'))])

#Listado de papeles como protagonista (n=1) que tuvo el actor "Charlton Heston" en la década de los 60's, ordenado por año de forma descendente
print(df_elenco[((df_elenco['name'] == 'Charlton Heston') & (df_elenco['n'] == 1) & (df_elenco['year'] >= 1960) & (df_elenco['year'] <= 1969))].sort_values(by = ['year']))

#Mostrar cuantos papeles para actores hubo en la década de los 50's
print(df_elenco[((df_elenco['year'] >= 1950) & (df_elenco['year'] <= 1959))].shape[0])
```

```python
print(f"Ha habido {df_elenco[((df_elenco['year'] >= 1950) & (df_elenco['year'] <= 1959))].shape[0]} papeles en la década de los 50")
#Mostrar cuantos papeles para actores hubo en la década de los 50's
print(f"{df_elenco[((df_elenco['year'] >= 1950) & (df_elenco['year'] <= 1959) & (df_elenco['type'] == 'actor'))].shape[0]} papeles han sido interpretados por actores y ")
#Mostrar cuantos papeles para actrices hubo en la década de los 50's
print(f"{df_elenco[((df_elenco['year'] >= 1950) & (df_elenco['year'] <= 1959) & (df_elenco['type'] == 'actress'))].shape[0]} por actrices")
```
