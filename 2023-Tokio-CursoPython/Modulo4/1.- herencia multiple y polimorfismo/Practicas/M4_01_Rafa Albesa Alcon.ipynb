**1) Ejercicio de herencia múltiple y polimorfismo:**
* Crea la clase Vehiculos con un constructor que incluya marca y modelo
* Esta clase Vehiculos también tendrá que incluir un método que se llame repostar() el cual imprima por pantalla "Este vehiculo tiene que repostar gasolina"
* Crea la clase VElectricos con un constructor que incluya marca, modelo y autonomia
* Esta clase VElectricos también tendrá que incluir un método que se llame repostar() el cual imprima por pantalla "Este vehiculo tiene que repostar electricidad"
* Crea la clase BicicletaElectrica que herede de Vehiculos y de VElectricos, pero dando prioridad a VElectricos (ya que es un vehículo más electrico que normal)
* Crea la clase Quad que herede de Vehiculos y de VElectricos, pero dando prioridad a Vehiculos (ya que sólo usa la electricidad de modo puntual)
* Crear un objeto de BicicletaElectrica y otro de Quad
* Emplear las técnicas de polimorfismo aprendidas para conseguir que al peguntarle a los objetos que acabamos de crear por su repostaje, ambos nos respondan adecuadamente



```python
class Vehiculo():
    __marca = ''
    __modelo = ''
    def __init__(self, marca, modelo):
        self.__marca = marca
        self.__modelo = modelo
    
    #*Getters:
    @property
    def marca(self):
        return self.__marca    
    
    @property
    def modelo(self):
        return self.__modelo
    
    #*Setters
    @marca.setter   
    def marca(self, valor):
        self.__marca = valor
    
    @modelo.setter   
    def modelo(self, valor):
        self.__modelo = valor
    
    #*Métodos operativos de la clase vehiculo
    def repostar(self):
        print("Este vehículo tiene que repostar gasolina")
        
    #*Overriden str method    
    def __str__(self):
        return f"Vehiculo marca {self.__marca} modelo {self.__modelo}"
    
    
class VElectrico():
    __marca = ''
    __modelo = ''
    __autonomia = 100
    def __init__(self, marca, modelo, autonomia):
        self.__marca = marca
        self.__modelo = modelo
        self.__autonomia = autonomia
    
    #*Getters:
    @property
    def marca(self):
        return self.__marca    
        
    @property
    def modelo(self):
        return self.__modelo
    
    @property
    def autonomia(self):
        return self.__autonomia
        
    #*Setters
    @marca.setter   
    def marca(self, valor):
        self.__marca = valor
        
    @modelo.setter   
    def modelo(self, valor):
        self.__modelo = valor
        
    @autonomia.setter   
    def autonomia(self, valor):
        self.__autonomia = valor
        
    #*Métodos operativos de la clase vehiculo
    def repostar(self):
        print("Este vehículo tiene que repostar electricidad")
    
    
    #*Overriden str method    
    def __str__(self):
        return f"Vehiculo eléctrico marca {self.__marca} modelo {self.__modelo} y una autonmía de {self.__autonomia} kilómetros"
    
class BicicletaElectrica(VElectrico, Vehiculo):
    def repostar(self):
        VElectrico.repostar(self)
        
class Quad(Vehiculo, VElectrico):
    def repostar(self):
        Vehiculo.repostar(self)
```

```python
vehiculo = Vehiculo('seat', '600')
print(vehiculo)
vehiculo.marca = 'Simca'
vehiculo.modelo ='1000'
print(vehiculo)
print(f"marca: {vehiculo.marca}")
print(f"modelo: {vehiculo.modelo}")
vehiculo.repostar()
velect = VElectrico('seat', '600', 300)
print(velect)
velect.marca = 'Simca'
velect.modelo ='1000e'
velect.autonomia = 300
print(velect)
print(f"marca: {velect.marca}")
print(f"modelo: {velect.modelo}")
print(f"autonomia: {velect.autonomia}")
velect.repostar()
print("-"*50)
bici = BicicletaElectrica('BH', '101','60')
bici.repostar()
print("-"*50)
quad = Quad('Montesa', '1250')
quad.repostar()
```

* Crea la superclase Personal_Universitario con un único atributo que sea un diccionario que incluya: "id", "nombre" y "email"

* Mostrar la información de esos objetos y comprobar que sale lo que tiene que salir

```python
class Personal_Universitario():
    __dict_datos = {'id': '', 'nombre':'', 'email':'' }
    
    def __init__(self, dict_datos):
        self.__dict_datos = dict_datos
        
    #*Getters
    @property
    def id(self):
        return self.__dict_datos

    #*Setters
    @id.setter
    def id(self, valor):
        self.__dict_datos = valor
        
    @property
    def dict_datos(self):
        return self.__dict_datos
    
    @dict_datos.setter
    def dict_datos(self, diccionario):
        self.__dict_datos = diccionario
    
    def __str__(self):
        return f"Persona de id {self.__dict_datos['id']}, nombre {self.__dict_datos['nombre']} y con email {self.__dict_datos['email']}"

class Oficina(Personal_Universitario):
    def __init__(self, dict_datos, puesto):
        dict_datos.update({'Puesto':puesto})
        super().__init__(dict_datos)

    def __str__(self):
        return f"Oficinista con id {super().dict_datos['id']}, nombre {super().dict_datos['nombre']} y con email {super().dict_datos['email']} y su puesto es {super().dict_datos['Puesto']}"
        
class Profesor(Personal_Universitario):
    def __init__(self, dict_datos, especializacion):
        dict_datos.update({'Especializacion':especializacion})
        super().__init__(dict_datos)

    def __str__(self):
        return f"Profesor con id {super().dict_datos['id']}, nombre {super().dict_datos['nombre']} y con email {super().dict_datos['email']} y está especializado en  {super().dict_datos['Especializacion']}"
    
class Alumno(Personal_Universitario):
    def __init__(self, dict_datos, creditos_aprobados):
        dict_datos.update({'Creditos aprobados':creditos_aprobados})
        super().__init__(dict_datos)

    def __str__(self):
        return f"Alumno con id {super().dict_datos['id']}, nombre {super().dict_datos['nombre']} y con email {super().dict_datos['email']} y tiene {super().dict_datos['Creditos aprobados']} créditos aprobados"
        
    

```
